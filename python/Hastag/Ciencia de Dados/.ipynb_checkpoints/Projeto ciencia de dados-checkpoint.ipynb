- TV, Jornal e Rádio estão em milhares de reais

#### Importar a Base de dados

```python
import pandas as pd

tabela = pd.read_csv("advertising.csv") #arquivo csv
display(tabela)
```

#### Análise Exploratória
- Vamos tentar visualizar como as informações de cada item estão distribuídas
- Vamos ver a correlação entre cada um dos itens

```python
import seaborn as sns #cria gráficos mas foi criado usando o matplotlib e tem gráficos mais fáceis para criar
import matplotlib.pyplot as plt #cria gráficos, exibe o gráfico

sns.heatmap(tabela.corr(), annot=True, cmap="Wistia") #criar gráfico, nome do gráfico, correlação das informações da tabela, exibir anotações , escala de cor
plt.show() #exibir

# outra forma de ver a mesma análise
# sns.pairplot(tabela)
# plt.show()
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
from sklearn.model_selection import train_test_split #biblioteca scikit-learn, pegando apenas a parte que eu quero usar dessa biblioteca

y = tabela["Vendas"] #quem eu quero prever
x = tabela.drop("Vendas", axis=1) #o resto, vou usar para fazer a previsão

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) #parametro para definir os dados de teste 30%, embaralhe aleatóriamente mas de forma fixa
#o padrão é 20%
#passar nessa ordem pq foi criado assim                 
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression #impostando um modelo de inteligencia artificial
from sklearn.ensemble import RandomForestRegressor #impostando outro modelo de inteligencia artificial

# cria as inteligencias aritificiais
modelo_regressaolinear = LinearRegression() #pega o modelo, cria
modelo_arvoredecisao = RandomForestRegressor() #pega o modelo, cria

# treina as inteligencias artificias, é sempre necessário
modelo_regressaolinear.fit(x_train, y_train) #treinar a inteligência artificial
modelo_arvoredecisao.fit(x_train, y_train) #treinar a inteligência artificial
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
from sklearn import metrics #utilizar as métricas da biblioteca

# criar as previsoes
previsao_regressaolinear = modelo_regressaolinear.predict(x_test) 
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_test)

# comparar os modelos
print(metrics.r2_score(y_test, previsao_regressaolinear)) #calcula a comparação entre a previsão da regressão linear e o teste
print(metrics.r2_score(y_test, previsao_arvoredecisao))  #calcula a comparação entre a previsão da arvore de decisão e o teste
```

#### Visualização Gráfica das Previsões

```python
#Árvore de decisão é o melhor modelo
#compararando as duas previsões no gráfico
tabela_auxiliar = pd.DataFrame() #criar uma tabela vazia
tabela_auxiliar["y_test"] = y_test #adicionar uma coluna
tabela_auxiliar["Previsoes ArvoreDecisao"] = previsao_arvoredecisao #adicionar uma coluna
tabela_auxiliar["Previsoes Regressao Linear"] = previsao_regressaolinear #adicionar uma coluna

plt.figure(figsize=(15,6)) #aumentar o tamanho da figura do gráfico
sns.lineplot(data=tabela_auxiliar) #criar um gráfico de linha
plt.show() #exibir
```
